```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

 
endpoint = "https://9n00400.openai.azure.com/" #要替換
model_name = "gpt-5.2" #要替換
deployment = "gpt-5.2" #要替換

load_dotenv()
subscription_key = os.getenv("5.2_AZURE_OPENAI_API_KEY") #寫在自己的.env
api_version = "2024-12-01-preview" #要替換

client = AzureOpenAI(
   api_version=api_version,
   azure_endpoint=endpoint,
   api_key=subscription_key,
#    http_client=http_client,
)

response = client.chat.completions.create(
   messages=[
       {
           "role": "system",
           "content": "You are a helpful assistant.",
       },
       {
           "role": "user",
           "content": "I am going to Paris, what should I see?",
       }
   ],
   max_completion_tokens=16384,
   model=deployment
)

print(response.choices[0].message.content)
```

```python
import os
from google import genai
from dotenv import load_dotenv

load_dotenv()

client = genai.Client(
    api_key=os.getenv("GEMINI_API_KEY")
)

response = client.models.generate_content(
    model="gemini-3.1-pro-preview",
    contents="說個笑話"
)

print(response.text)
```

```python
from anthropic import AnthropicFoundry
import os

endpoint = "https://project3-docai-resource.services.ai.azure.com/anthropic/"
deployment_name = "claude-opus-4-5"
api_key = os.getenv("AZURE_ANTHROPIC_API_KEY")

client = AnthropicFoundry(
   api_key=api_key,
   base_url=endpoint,
#    http_client=http_client,
)

message = client.messages.create(
   model=deployment_name,
   messages=[
       {"role": "user", "content": "What is the capital of France?"}
   ],
   max_tokens=1024,
)

print(message.content[0].text)
```
